```python
from tkinter import * 
from tkinter import messagebox
   
master = Tk()                                                                   #main window declaration
  
# sets the geometry of main  window
master.geometry("300x300") 

#########################################################################

def check(name):                                                                #to ensure only alphabets
    f=0
    for i in range(0,len(name)):
        if(not((ord(name[i])>64 and ord(name[i])<91)or(ord(name[i])>96 and ord(name[i])<123)or(name[i]==" "))):
            f=1
    if(f==0):
        if(name.isspace()==True):
            f=2
    return f
#########################################################################

def message(t,m):                                                                #message box
    if(m==1):
        m="Contact Book Created"
    elif(m==2):
        m='Contact Book Already Exist'
    elif(m==3):
        m="Phone and social security number will be all digit try again"
    elif(m==4):
        m="Check the length of phone and social security number again"
    elif(m==5):
        m="Sucessfully added"
    elif(m==6):
        m='File does not exists, please try a different name'
    elif(m==7):
        m="Name should only have alphabets"
    elif(m==8):
        m="Name too long, please use initial"
    elif(m==9):
        m="Field couldn't be blank"
    elif(m==10):
        m="Address is too long try in brief or short forms"
        
    if(t==0):
        t="Sucess"
    elif(t==1):
        t="Failure"
    elif(t==2):
        t="Warning"
    messagebox.showinfo(t,m)

########################################################################

def NewBook():                                                             #view window
      
    # Toplevel object which will  
    # be treated as a new window 
    newWindow = Toplevel(master) 
  
    newWindow.title("New Contact Book") 
  
    newWindow.geometry("400x200") 
  
    Label(newWindow,text ="This window is for creating new contact book").grid(row=1,column=1,pady=20)
    
    x=StringVar()
    def create():
        n=x.get()
        n=n+".txt"
        try:
            with open(n,'x') as file:
                list1=["Sl. No.","Phone No. ","SocSecNo.","Name","Address"]
                for i in list1:
                    if(i=="Sl. No."):
                        file.write(i)
                        l=12-len(i)
                        for j in range(0,l):
                            file.write(" ")
                    elif(i=="Name"):
                        file.write(i)
                        l=32-len(i)
                        for j in range(0,l):
                            file.write(" ")
                    else:
                        file.write(i)
                        file.write(" ")
                file.write("\n")
                message(0,1)
                #print("Contact Book Created")
        except FileExistsError:
            message(1,2)
            #print('Contact Book Already Exist')
    
    Label(newWindow,text ="Enter contact book name").grid(row=2,column=1)
    e1=Entry(newWindow,textvariable=x).grid(row=2,column=2,pady=10)
    b1=Button(newWindow,text="Submit",command=create).grid(row=3,column=2)
    
############################################################################

def NewContact():                                   #for adding new contact
    newWindow = Toplevel(master) 
  
    newWindow.title("New Contact") 

    newWindow.geometry("400x200") 
  
    Label(newWindow,text ="This window is for new contact").grid(row=1,column=1,pady=20)
    
    x=StringVar()
    def AddContact():                                                                 #adding contact begin
        n=x.get()
        n=n+".txt"
        try:
            with open(n,'r') as file:
                file.close()
                newWindow1 = Toplevel(master) 
                newWindow1.title("New Contact") 
                newWindow1.geometry("400x200")
                
                
                name_var=StringVar()
                phone_var=StringVar()
                ssn_var=StringVar()
                address_var=StringVar()
                
                def details():                                                     #enter details for contact
                    name=name_var.get()
                    phone=phone_var.get()
                    ssn=ssn_var.get()
                    address=address_var.get()
                    
                    if(check(name)>0):
                        if(check(name)==1):
                            message(2,7)
                        if(check(name)==2):
                            message(2,9)
                    elif(len(name)>=30):
                        message(2,8)
                    elif(phone.isdigit()!=True or ssn.isdigit()!=True):
                        if(len(phone)==0 or len(ssn)==0):
                            message(2,9)
                        else:
                            message(2,3)
                    elif(len(phone)!=10 or len(ssn)!=9):
                        message(2,4)
                    elif(len(address)==0 or address.isspace==True):
                        message(2,9)
                    elif(len(address)>=60):
                        message(2,10)
                    else:
                        f=open(n,'r')
                        content=f.read()
                        CoList = content.split("\n") 
                        LineCounter=0
                        for i in CoList: 
                            if i: 
                                LineCounter=LineCounter+ 1
                        f.close()
                        f=open(n,'a')
                        list2=[LineCounter,phone,ssn,name,address]
                        for i in range(0,5):
                            if(i==0):
                                f.write(str(list2[i]))
                                l=str(list2[i])
                                l=12-len(l)
                                for j in range(0,l):
                                    f.write(" ")
                            elif(i==3):
                                f.write(str(list2[i]))
                                l=str(list2[i])
                                l=32-len(l)
                                for j in range(0,l):
                                    f.write(" ")
                            else:
                                f.write(str(list2[i]))
                                f.write(" ")
                        f.write("\n")
                        f.close()
                        message(0,5)
                
                newWindow1.title("Fill Details")
                newWindow1.geometry("300x200")
                Label(newWindow1,text ="Enter Name").grid(row=2,column=1)
                Label(newWindow1,text ="Enter Phone Number").grid(row=3,column=1)
                Label(newWindow1,text ="Enter SSN").grid(row=4,column=1)
                Label(newWindow1,text ="Enter Address").grid(row=5,column=1)
                e1=Entry(newWindow1,textvariable=name_var).grid(row=2,column=2,pady=10)
                e2=Entry(newWindow1,textvariable=phone_var).grid(row=3,column=2,pady=10)
                e3=Entry(newWindow1,textvariable=ssn_var).grid(row=4,column=2,pady=10)
                e4=Entry(newWindow1,textvariable=address_var).grid(row=5,column=2,pady=10)
                
                b1=Button(newWindow1,text="submit",command=details).grid(row=7,column=2)
                
        except FileNotFoundError:
            message(1,6)
            
    Label(newWindow,text ="Enter address book name you want to add to").grid(row=2,column=1)
    e1=Entry(newWindow,textvariable=x).grid(row=2,column=2,pady=10)
    b1=Button(newWindow,text="search",command=AddContact).grid(row=3,column=2)
    
###############################################################################################

def View():                                                             #view window
      
    # Toplevel object which will  
    # be treated as a new window 
    newWindow = Toplevel(master) 
  
    newWindow.title("VIEW") 
   
    newWindow.geometry("400x200") 
   
    Label(newWindow,text ="This window is for searching and viewing").grid(row=1,column=1,pady=20)
    
    x=StringVar()
    def disp():                                                                 #opening file
        n=x.get()
        n=n+".txt"
        try:
            with open(n,'r') as file:
                print(file.read())
        except FileNotFoundError:
            message(1,6)
            
    Label(newWindow,text ="Enter address book name").grid(row=2,column=1)
    e1=Entry(newWindow,textvariable=x).grid(row=2,column=2,pady=10)
    b1=Button(newWindow,text="search",command=disp).grid(row=3,column=2)
    
#############################################################################

def Add():                                                             #Add window
      
    # Toplevel object which will  
    # be treated as a new window 
    newWindow = Toplevel(master) 
  
    newWindow.title("Add Window") 
   
    newWindow.geometry("300x300") 
   
    Label(newWindow,text ="This window is for adding").pack(pady=10) 
    
    btn1 = Button(newWindow, text ="Create new contact book", command=NewBook) 
    btn1.pack(pady = 10,ipady=20,ipadx=20) 
    btn2 = Button(newWindow, text ="Add contact to existing", command=NewContact) 
    btn2.pack(pady = 20,ipady=20,ipadx=20) 
    
#############################################################################

master.title("Address Generator")
label = Label(master,text ="This is the main window") 
  
label.pack(pady = 10) 
  

mw_btn1 = Button(master, text ="VIEW CONTACT",command = View) 
mw_btn1.pack(pady = 10,ipady=20,ipadx=20) 

mw_btn2 = Button(master, text ="ADD",command = Add) 
mw_btn2.pack(pady = 20,ipady=20,ipadx=20) 
  
mainloop() 
```
